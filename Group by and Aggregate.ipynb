# Group by and Aggregate

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
group_by_frame.mean()
```

```python
df.groupby('Base Flavor').mean()
```

```python
 df.groupby('Base Flavor').max()
```

```python
 df.groupby('Base Flavor').sum()
```

```python
 df.groupby('Base Flavor').agg({'Flavor Rating':['mean','max','min','sum'],'Texture Rating':['mean','max','min','sum']})
```
