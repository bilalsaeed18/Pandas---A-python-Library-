# Data Visualization in Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Bilal Home\Documents\Projects\Pandas-A-python-library\Dataset\Flavors.csv")
df
```

```python
print(plt.style.available)
plt.style.use('dark_background')
```

```python
df.plot()
```

```python
df.plot( kind = 'line')
```

```python
df.plot( kind = 'line', subplots = True)
```

```python
df.plot( kind = 'line', title = 'Ratings', xlabel = 'Rating', ylabel = 'Scores')
```

```python
df.plot( kind = 'bar', title = 'Ratings')
```

```python
df.plot( kind = 'bar', stacked = True, title = 'Ratings')
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.hist(bins = 20)
```

```python
df.boxplot()
```

```python
df.plot.area()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y = 'Flavor Rating')
```
